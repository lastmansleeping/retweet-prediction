```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")
words = set()
w = []
for s in y:
    #print s
    s = re.sub('@[a-zA-Z0-9_]+','', s)
    s = re.sub('(http|https):[a-zA-Z0-9/]+','',s)
    w.append(s)
    tokenizer = nltk.RegexpTokenizer(r'\w+')
    x = tokenizer.tokenize(s)
    if "RT" in x:
        x = x[1:]
    
    for word in x:
        word = stemmer.stem(word)
        words.add(word.lower())
    
       
```

```python
row_iterator = df.iterrows()
for i,row in row_iterator: 
    s = row['tweet']
    s = re.sub('@[a-zA-Z0-9_]+','', s)
    s = re.sub('(http|https):[a-zA-Z0-9/]+','',s)
    w.append(s)
    tokenizer = nltk.RegexpTokenizer(r'\w+')
    x = tokenizer.tokenize(s)
    if "RT" in x:
        x = x[1:]
    y = []
    for word in x:
        y.append(stemmer.stem(word))
    
    string = ' '.join(y)
    df.ix[i, 'tweet'] = string
```

```python
clf = Pipeline([('vect', TfidfVectorizer(stop_words=sorted(nltk.corpus.stopwords.words('english')))),('clf', MultinomialNB())])
```

```python
clf.fit(df['tweet'], df['retweet'])
```

```python
from sklearn import metrics

print "Accuracy : "
print clf.score(df['tweet'], df['retweet'])
    
y_pred = clf.predict(df['tweet'])
    
print "Classification Report:"
print metrics.classification_report(df['retweet'] , y_pred)
print "Confusion Matrix:"
print metrics.confusion_matrix(df['retweet'], y_pred)
```

```python
from sklearn import metrics
def plot_roc(algo_name, targets, scores, ph):
    fpr, tpr, thresholds = metrics.roc_curve(targets, scores)
    roc_auc = metrics.auc(fpr, tpr)
    print algo_name
    ph.plot(fpr, tpr, label=str(algo_name) + '(%0.2f)' % roc_auc)
    return roc_auc
```

```python
from sklearn import metrics
def plot_roc(algo_name, targets, scores, ph):
    fpr, tpr, thresholds = metrics.roc_curve(targets, scores)
    roc_auc = metrics.auc(fpr, tpr)
    fig = ph.figure()
    ax = fig.add_subplot(111)
    fig.subplots_adjust(top=1)
    ax.set_title('ROC Curve')
    ax.text(0.95, 0.01, 'Area Under Curve : (%0.2f)' % roc_auc,
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='green', fontsize=15)
    ph.plot(fpr, tpr, label=str(algo_name) + '(%0.2f)' % roc_auc)
    ph.show()
    #return roc_auc

   
plot_roc('Naive Bayes', df['actual'], df['probability'], plt)
```
