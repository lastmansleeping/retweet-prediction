```python
#Finding username

def username():
    nouns = ["NN","NNS","NNP","NNPS"]
    row_iterator = df_tweets.iterrows()
    for i,row in row_iterator:
        flag = 0
        tokens = nltk.word_tokenize(str(row['tweet']))
        if row['retweet'] == 1:
            tokens = tokens[4:]

        pos_tokens = nltk.pos_tag(tokens)
        if "@" in tokens:
            index = tokens.index("@")
            try:
                if pos_tokens[index + 1][1] in nouns:
                    flag = 1
            
            except:
                pass
                
        if flag == 1:
            df_tweets['username'][df_tweets['tweet'] == row['tweet']] = 1

            
```

```python
df_tweets['username'] = 0
username()
```

```python
#hashtag
import re

def hashtag():
    row_iterator = df_tweets.iterrows()
    for i,row in row_iterator:
        flag = 0
        hashtags = re.findall(r"#(\w+)", row['tweet'])
        if len(hashtags) > 0:
            df_tweets['hashtag'][df_tweets['tweet'] == row['tweet']] = 1
```

```python
#url

def url():
    row_iterator = df_tweets.iterrows()
    for i,row in row_iterator:
        urls = re.findall(r"http:", row['tweet'])
        if len(urls) > 0:
            df_tweets['url'][df_tweets['tweet'] == row['tweet']] = 1
```

```python
#exclamation and question

def stage3():
    row_iterator = df_tweets.iterrows()
    for i,row in row_iterator:
        if re.search(r"!", row['tweet']):
            df_tweets['exclamation'][df_tweets['tweet'] == row['tweet']] = 1
        
        tokens = nltk.word_tokenize(row['tweet'])
        if "?" in tokens:
            df_tweets['question'][df_tweets['tweet'] == row['tweet']] = 1
        
        
```

```python
#positive term and negative term

def stage4():
    pos_words = ["great","like","excellent","rock on"]
    neg_words = ["f**k","suck","fail","eww"]
    pos_emoticon = [":-)",":)",";-)"]
    neg_emoticon = [":-(",":("]
    row_iterator = df_tweets.iterrows()
    for i,row in row_iterator:
        for word in pos_words:
            if word in row['tweet'].lower():
                df_tweets['positive_term'][df_tweets['tweet'] == row['tweet']] = 1
          
        for word in neg_words:
            if word in row['tweet'].lower():
                df_tweets['negative_term'][df_tweets['tweet'] == row['tweet']] = 1
                
        for word in pos_emoticon:
            if word in row['tweet'].lower():
                df_tweets['positive_emoticon'][df_tweets['tweet'] == row['tweet']] = 1
                
        for word in neg_emoticon:
            if word in row['tweet'].lower():
                df_tweets['negative_emoticon'][df_tweets['tweet'] == row['tweet']] = 1
                
    
```

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")
row_iterator = df_anew.iterrows()
for i,row in row_iterator:
    df_anew['Word'][df_anew['Word'] == row['Word']] = stemmer.stem(row['Word'])
```

```python
#stage 5

def stage5():
    row_iterator = df_tweets.iterrows()
    for i,row in row_iterator:
        valence = []
        arousal = []
        dominance = []
        tokens = nltk.word_tokenize(row['tweet'])
        stemmed_tokens = [stemmer.stem(t) for t in tokens]
        #print stemmed_tokens
   
        for t in stemmed_tokens:
            if len(df_anew[df_anew['Word'] == t]) > 0:
                row_entry = df_anew[df_anew['Word'] == t]
                #print
                #print row['V.Mean.Sum'].values[0]
                #print t
                valence.append(row_entry['V.Mean.Sum'].values[0])
                arousal.append(row_entry['A.Mean.Sum'].values[0])
                dominance.append(row_entry['D.Mean.Sum'].values[0])
        
        if len(valence) > 0:
            valence = sum(valence)/len(valence) #calculating average value
            arousal = sum(arousal)/len(arousal)
            dominance = sum(dominance)/len(dominance)
                    
            valence = valence - 5
            arousal = arousal - 5
            dominance = dominance - 5
                    
        else:
            valence = 0
            arousal = 0
            dominance = 0
            
        df_tweets['valence'][df_tweets['tweet'] == row['tweet']] = valence
        df_tweets['arousal'][df_tweets['tweet'] == row['tweet']] = arousal      
        df_tweets['dominance'][df_tweets['tweet'] == row['tweet']] = dominance    
                
                
                
```

```python
stage5()
```
